# <center>Sakila Dataset</center>

Using your `Sakila` [Sakila Reference](https://dev.mysql.com/doc/sakila/en/sakila-structure.html) schema as the base is a strong choice. The official Sakila documentation shows the sample database includes 16 base tables, 7 views, 3 stored procedures, 3 stored functions, and 6 triggers

---
### Solve in this order:
1.	Basic retrieval and filtering
2.	Sorting, grouping, joins
3.	Subqueries and set logic
4.	CTEs and window functions
5.	DML and transactions
6.	Views, procedures, functions, triggers
7.	Indexing, optimization, and design
8.	Advanced reporting and real interview scenarios
---

```ini
A good rule is:
	•	first solve with a direct query
	•	then rewrite with joins
	•	then rewrite with subqueries or CTEs
	•	then optimize it
	•	then explain edge cases
```

## <center> 1. Basic `SELECT`questions </center>

1.	List all customers.
```sql
SELECT * 
FROM sakila.customer
```

2.	Show only customer first name, last name, and email.
```sql
SELECT first_name, last_name, email
FROM sakila.customer
```

3.	List all films with title and release year.
```sql
SELECT title, release_year
FROM sakila.film
```

4.	Show all active customers.
```sql
SELECT active
FROM sakila.customer
WHERE active = 1
```

5.	Find all staff members.
```sql
SELECT *
FROM sakila.staff
```

6.	Show all countries in alphabetical order.
```sql
SELECT country
FROM sakila.country
ORDER BY country DESC
```

7.	List all cities with their country IDs.
```sql
SELECT city, country_id
FROM sakila.city
```

8.	Display all categories.
```sql
SELECT *
FROM sakila.category
```

9.	Show all languages available in the database.
```sql
SELECT name
FROM sakila.language
```

10.	List all stores.
```sql
SELECT *
FROM sakila.stores
```

11.	Show all inventory rows.
```sql
SELECT *
FROM sakila.stores
```

12.	List all rentals.
```sql
SELECT *
FROM sakila.rentals
```

13.	Display all payments.
```sql
SELECT *
FROM sakila.payment
```

14.	Show all actors.
```sql
SELECT *
FROM sakila.actor
```

15.	List film titles and rental rates.
```sql
SELECT title, rental_rate
FROM sakila.film
```

17.	List customers created after a specific date.
```sql
SELECT customer_id, first_name, last_name, create_date
FROM customer
WHERE create_date >= '2006-01-09'
```

18.	Show payments made on a specific date.
```sql
SELECT payment_id, customer_id, payment_date
FROM payment
WHERE DATE(payment_date) = '2006-02-14' -- ignore the time part and only validate the date
```

19.	List rentals where return_date is null.
```sql
SELECT rental_date, return_date
FROM rental
WHERE rental_date is NULL
```

20.	Show customers whose email is null or missing.
```sql
SELECT customer_id, first_name, last_name, email
FROM customer
WHERE email is NULL OR email = " "
```

21.	List films where description is null.
```sql
SELECT film_id, title, description
FROM film
WHERE description = " " OR description is NULL
```

22.	Show films where length is null.
```sql
SELECT film_id, title, description
FROM film
WHERE length = 0 OR length is NULL
```

24.	List addresses with postal codes.
```sql
SELECT address
FROM address
WHERE postal_code is NOT NULL
```

25.	Show all district names from address.
```sql
SELECT district
FROM address
WHERE district = '' OR district is NULL
```

## <center>3. `ORDER BY`, `LIMIT`, `DISTINCT`</center>

```sql
SELECT *
FROM customer
ORDER BY last_name
```

```sql
SELECT *
FROM film
ORDER BY `length` DESC
```

* use `TOP` only when getting raw DATA
* If you're filtering the data you need to use `LIMIT`
```sql
SELECT *
FROM payment
ORDER BY amount DESC
LIMIT 10
```

```sql
SELECT *
FROM film
ORDER BY rental_rate ASC
LIMIT 5
```

```sql
SELECT DISTINCT rating
FROM film
```

```sql
SELECT DISTINCT store_id
FROM store
```

```sql
SELECT *
FROM actor
ORDER BY first_name ASC, last_name ASC
```
9.	Show films ordered by release year and title.
```sql
SELECT *
FROM film
ORDER BY release_year DESC, title ASC
```
10.	Find top 5 longest films.
```sql
SELECT *
FROM film
ORDER BY rental_duration DESC
LIMIT 5
```
11.	Show bottom 10 payments by amount.
```sql
SELECT *
FROM payment
ORDER BY amount ASC
LIMIT 10
```
12.	Display customers sorted by create_date descending.
```sql
SELECT *
FROM customer
ORDER BY create_date DESC
```
13.	Find distinct replacement_cost values.
```sql
SELECT DISTINCT(replacement_cost)
FROM film
```
14.	Show last 20 rentals by rental_date.
```sql
SELECT *
FROM rental
ORDER BY rental_date DESC
LIMIT 20
```
15.	List unique categories.
```sql
SELECT DISTINCT(category_id), name, last_update
FROM category
```
16.	Show titles alphabetically with limit 15.
```sql
SELECT *
FROM film
ORDER BY title
LIMIT 15
```
17.	Find first 10 customers from store 1 ordered by first name.
```sql
SELECT *
FROM customer
WHERE store_id=1
ORDER BY first_name
```
18.	Show films ordered by rating then rental_duration.
```sql
SELECT *
from film
ORDER BY rating, rental_duration
```

```sql
SELECT COUNT(DISTINCT(postal_code))
FROM address
```
20.	Display unique city IDs used in address.
```sql
SELECT DISTINCT(city_id), district
FROM address
ORDER BY city_id
```
