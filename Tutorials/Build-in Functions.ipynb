## `CHAR_LENGTH()`
* The `CHAR_LENGTH()` function return the length of a string (in characters).

```sql
SELECT tweet_id
FROM Tweets
WHERE CHAR_LENGTH(content) > 15
```
---

## `LENGTH()`
* The `LENGTH()` function returns the length of a string (in bytes).

```sql
SELECT LENGTH(CustomerName) AS LengthOfName
FROM Customers;
```
