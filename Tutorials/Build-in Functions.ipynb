## `CHAR_LENGTH()`
* The `CHAR_LENGTH()` function return the length of a string (in characters).

```sql
SELECT tweet_id
FROM Tweets
WHERE CHAR_LENGTH(content) > 15
```
---

## `LENGTH()`
* The `LENGTH()` function returns the length of a string (in bytes).

```sql
SELECT LENGTH(CustomerName) AS LengthOfName
FROM Customers;
```
---
## `COUNT()`
* The `COUNT(col name)` counts all non-null values in the specified columns.

```sql
SELECT COUNT(ProductName)
FROM Products;
```
- `COUNT(DISTINCT col_name)` counts only the unique, non-null values in the column. 

```sql
-- Calculate the number of unique subjects each teacher teaches in the university
SELECT teacher_id, COUNT(DISTINCT subject_id) as cnt
FROM Teacher
GROUP BY teacher_id;
```

- Add `WHERE` clause to specify conditions

```sql
-- Count the number of products where Price is higher than 20
SELECT COUNT(ProductID)
FROM Products
WHERE Price > 20;
```

- We can use `COUNT()` function and the `GROUP BY` clause, to return the number of EACH category in the 'PRODUCTS' table.

```sql
SELECT COUNT(*) AS [Number of records], CategoryID
FROM Products
GROUP BY CategoryID;
```
```sql
-- For each date_id & make_name, find the number of distinct lead_id and distinct partner_id
SELECT date_id, make_name, COUNT(DISTINCT lead_id) as unique_leads, COUNT(DISTINCT partner_id) as unique_partners
FROM DailySales
GROUP BY date_id, make_name;
```
