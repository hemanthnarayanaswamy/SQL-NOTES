# <center> SQL JOINS </center>

* The `JOIN` clause is used to combine rows from two or more tables.

[Practice Joins](https://learnsql.com/blog/sql-join-examples-with-explanations/)

![intro](../Resources/assets/joins/1.png)

1. `(INNER) JOIN`: Returns only rows that have **matching values** in both tables
2. `LEFT (OUTER) JOIN`: Returns **all rows from the left table**, and only the matched rows from the `right table`
3. `RIGHT (OUTER) JOIN`: Returns **all rows from the right table**, and only the matched rows from the `left table`
4. `FULL (OUTER) JOIN`: Returns **all rows** when there is a match in either the left or right table
