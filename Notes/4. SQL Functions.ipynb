# <center> SQL Functions </center>
- SQL functions are built-in or user-defined operations that perform calculations and manipulate data, returning a result to simplify queries and data analysis. 

### Types of SQL Functions
1. `Aggregate Functions`: Operate on sets of values to return a single value, often used with `GROUP BY`
Example: `AVG(), COUNT(), MAX(), MIN(), SUM()`

2. `SCALAR FUNCTIONS`: Perform operations on individual values. Key types include string manipulation `CONCAT, LENGTH, UPPER`, mathemcatical calculations `ABS, ROUND` and date/time handling `NOW, DATEDIFF`

3. `WINDOW FUNCTIONS`: Perform calculations across a specific set of rows related to the current row without collapsing them, such as `ROW_NUMBER(), RANK(), DENSE_RANK()`

4. `USER DEFINED FUNCTIONS (UDFs)`: Custom, reusable functions created for specific logic, which can return single values or tables.

### B. `DATEPART()`
- The `DATEPART()` functin in SQL extracts a specific part of a date or time and returns it as an **interger**. 
- **Syntax** `DATEPART(interval, date)` where `interval` specifies the part to extract and the `date` is the expression to evaluate.

![f12](../Resources/assets/sql_functions/f13.png)

```sql
SELECT DATEPART(year, GETDATE()) AS Year, 
       DATEPART(month, GETDATE()) AS Month, 
       DATEPART(day, GETDATE()) AS Day,
       DATEPART(week, GETDATE()) AS Week, 
       DATEPART(weekday, GETDATE()) AS Weekday, 
       DATEPART(hour, GETDATE()) AS Hour,
       DATEPART(minute, GETDATE()) AS Minute, 
       DATEPART(second, GETDATE()) AS Second, 
       DATEPART(quarter, GETDATE()) AS Quarter;  
```

### C. `DATENAME()`
- The `DATENAME()` function in sql extracts a specified part of a date or time value and return the result as a **STRING**.

```sql
-- Get the full weekday name
SELECT DATENAME(weekday, '2022-01-01'); -- Returns: 'Saturday'

-- Get the month name
SELECT DATENAME(month, '2020-03-05'); -- Returns: 'March'

-- Get the year as a string
SELECT DATENAME(year, '2025-12-31'); -- Returns: '2025'   
```

### E. `EOMONTH()`
- The `EOMONTH()` function returns the **last day** of the month for a specified date. 
- `EOMONTH(date, offset)` 

```sql
SELECT EOMONTH('2022-09-25') -- '2022-09-30'
SELECT EOMONTH('2022-09-25', 1) -- '2022-10-31'
SELECT EOMONTH('2022-09-25', -1) -- '2022-08-31'
```

### G. `CONVERT()`
- The `CONVERT()` function converts a value (of any type) into a specified datatype.

```sql
CONVERT(data_type(length), expression, style)

-- data_type: The datatype to convert expression to (bigint, int, smallint, bit,decimal, numeric, money, float, real, datetime, char, varchar, text, binary or image etc...)
-- length: optional. The length of the resulting data type
-- expression: The value to convert to another data type
```
