# <center> SQL Functions </center>
- SQL functions are built-in or user-defined operations that perform calculations and manipulate data, returning a result to simplify queries and data analysis. 

### Types of SQL Functions
1. `Aggregate Functions`: Operate on sets of values to return a single value, often used with `GROUP BY`
Example: `AVG(), COUNT(), MAX(), MIN(), SUM()`

2. `SCALAR FUNCTIONS`: Perform operations on individual values. Key types include string manipulation `CONCAT, LENGTH, UPPER`, mathemcatical calculations `ABS, ROUND` and date/time handling `NOW, DATEDIFF`

3. `WINDOW FUNCTIONS`: Perform calculations across a specific set of rows related to the current row without collapsing them, such as `ROW_NUMBER(), RANK(), DENSE_RANK()`

4. `USER DEFINED FUNCTIONS (UDFs)`: Custom, reusable functions created for specific logic, which can return single values or tables.

### B. `DATEPART()`
- The `DATEPART()` functin in SQL extracts a specific part of a date or time and returns it as an **interger**. 
- **Syntax** `DATEPART(interval, date)` where `interval` specifies the part to extract and the `date` is the expression to evaluate.

![f12](../Resources/assets/sql_functions/f13.png)

```sql
SELECT DATEPART(year, GETDATE()) AS Year, 
       DATEPART(month, GETDATE()) AS Month, 
       DATEPART(day, GETDATE()) AS Day,
       DATEPART(week, GETDATE()) AS Week, 
       DATEPART(weekday, GETDATE()) AS Weekday, 
       DATEPART(hour, GETDATE()) AS Hour,
       DATEPART(minute, GETDATE()) AS Minute, 
       DATEPART(second, GETDATE()) AS Second, 
       DATEPART(quarter, GETDATE()) AS Quarter;  
```

### C. `DATENAME()`
- The `DATENAME()` function in sql extracts a specified part of a date or time value and return the result as a **STRING**.

```sql
-- Get the full weekday name
SELECT DATENAME(weekday, '2022-01-01'); -- Returns: 'Saturday'

-- Get the month name
SELECT DATENAME(month, '2020-03-05'); -- Returns: 'March'

-- Get the year as a string
SELECT DATENAME(year, '2025-12-31'); -- Returns: '2025'   
```

# <center>4. NULL FUNCTIONS</center>

- `NULL` value represents an unknown or missing value. It is different from zero or an empty string. When working with `NULL` values in SQL, it is important to understand how to handle them correctly to avoid unexpected result.

1. Comparisons with `NULL` using regular comparison operators like `"=", "<>", "<", ">"` do not yield `true/false` but rather produce a result of unknown or `NULL`, you need to use the `IS NULL or IS NOT NULL` operators to check for `NULL` values.
2. When performing arithmetic or other operations involving `NULL` values, the result typically becomes `NULL`
3. Most aggregate functionsi n SQL, ignore `NULL` values when calculaitng results.

### A. `ISNULL()`
- The `ISNULL()` function is used to replace NULL values with the specified replacement values.

![n1](../Resources/assets/sql_functions/null/n1.png)

```sql
-- Find the sum of the salary of all Employees, if the Salary of any employee is not available (or NULL value), use salary as 10000.
SELECT SUM(ISNULL(Salary, 10000) AS Salary
FROM Employee;
```
---

### B. `COALESCE()`
- `COAlESCE` function in SQL returns the first non-NULL expression among its arguments. If all the expressions evaluate to null, then the COALESCE function will return null.

![n2](../Resources/assets/sql_functions/null/n2.png)

EXAMPLE: **FETCH the name and contact number of each employee.**
![sample](https://media.geeksforgeeks.org/wp-content/uploads/7-4.jpg)

```sql
SELECT Name, COALESCE(Phone1, Phone2) AS Contact
FROM Contact_info;
```

# <center>5. CASE STATEMENTS</center>

- The `CASE` expression is used to define different results based on specified conditions in an SQL statement.

- The `CASE` expression goes through the conditions and stops at the first match (like `if-elif-else` statement). So once a condition is true, it will stop processing and return the result. 
- If no conditions are true, it returns the value in the `ELSE` caluse and no conditions are true, it returns `NULL`.

![n6](../Resources/assets/sql_functions/null/n6.png)

**EXAMPLE**

```sql
-- This query evaluates the quantity of each order and returns a corresponding text message.
SELECT OrderID, Quantity,
   CASE
       WHEN Quantity > 30 THEN 'The quantity is greater than 30'
       WHEN Quantity = 30 THEN 'The quantity is 30'
       ELSE 'The quantity is under 30'
   END AS QuantityText
FROM OrderDetails;

-- This query ensures that customers with a NULL city are ordered by their country.
SELECT CustomerName, City, Country
FROM Customers
ORDER BY
   CASE
       WHEN City IS NULL THEN Country
       ELSE City
   END;

-- We can have the column in with CASE and define the values directly instead of equating them
SELECT 
    ProductName,
    CategoryID,
    CASE CategoryID
        WHEN 1 THEN 'Beverages'
        WHEN 2 THEN 'Condiments'
        ELSE 'Other'
    END AS CategoryName
FROM Products;
```
