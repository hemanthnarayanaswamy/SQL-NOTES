# <center> SQL JOINS </center>

* The `JOIN` clause is used to combine rows from two or more tables.

[Practice Joins](https://learnsql.com/blog/sql-join-examples-with-explanations/)

![intro](../Resources/assets/joins/1.png)

1. `(INNER) JOIN`: Returns only rows that have **matching values** in both tables
2. `LEFT (OUTER) JOIN`: Returns **all rows from the left table**, and only the matched rows from the `right table`
3. `RIGHT (OUTER) JOIN`: Returns **all rows from the right table**, and only the matched rows from the `left table`
4. `FULL (OUTER) JOIN`: Returns **all rows** when there is a match in either the left or right table

![intro](../Resources/assets/joins/2.png)

# <center>1. SQL INNER JOIN</center>
`INNER JOIN` is used to retrieve rows where matching values exist in both tables. It helps in:
* Combining records based on a related column.
* Returning only matching rows from both tables.
* Excluding non-matching data from the result set.
* Ensuring accrate data relationships between tables.

![innerJoin](../Resources/assets/joins/j1.png)

**SYNTAX**
```sql
SELECT table1.column1,table1.column2,table2.column1,.... 
FROM table1 INNER JOIN table2 ON  table1.matching_column = table2.matching_column; 
```

```sql
SELECT StudentCourse.COURSE_ID, Student.NAME, Student.AGE 
FROM Student 
INNER JOIN StudentCourse 
ON Student.ROLL_NO = StudentCourse.ROLL_NO;
```
---
# <center>2. SQL LEFT JOIN</center>
`LEFT JOIN` is used to retrieve all rows from the left table and matching rows from the right table. It helps in:
* Returning all records from the left table.
* Showing matching data from the right table.
* Displaying `NULL` values where no match exists in the right table.
* Performing outer joins, also known as `LEFT OUTER JOIN`

![LEFTJoin](../Resources/assets/joins/j2.png)

**SYNTAX**
```sql
SELECT table1.column1,table1.column2,table2.column1,....
FROM table1 
LEFT JOIN table2
ON table1.matching_column = table2.matching_column;
```

```sql
SELECT Student.NAME,StudentCourse.COURSE_ID 
FROM Student
LEFT JOIN StudentCourse 
ON StudentCourse.ROLL_NO = Student.ROLL_NO;
```
---
# <center>3. SQL RIGHT JOIN</center>
`RIGHT JOIN` is used to retrieve all rows from the right table and the matching rows from the left table. It helps in:
* Returning all records from the right-side table.
* Showing matching data from the left-side table.
* Displaying `NULL` values where no match exists in the left table.
* Performing outer joins, also known as `RIGHT OUTER JOIN`

![rightJoin](../Resources/assets/joins/j3.png)

**SYNTAX**
```sql
SELECT table1.column1,table1.column2,table2.column1,....
FROM table1 
RIGHT JOIN table2
ON table1.matching_column = table2.matching_column;
```

```sql
SELECT Student.NAME,StudentCourse.COURSE_ID 
FROM Student
RIGHT JOIN StudentCourse 
ON StudentCourse.ROLL_NO = Student.ROLL_NO;
```
---
# <center>4. SQL FULL JOIN</center>
`FULL JOIN` is used to combine the results of both `LEFT JOIN` & `RIGHT JOIN`. It helps in:
* Returning all rows from both tables.
* Showing matching records from each table.
* Displaying `NULL` values where no match exists in either table.
* Providing complete data from both sides of the join.

![fullJoin](../Resources/assets/joins/j4.png)

**SYNTAX**
```sql
SELECT table1.column1,table1.column2,table2.column1,....
FROM table1 
FULL JOIN table2
ON table1.matching_column = table2.matching_column; 
```

```sql
SELECT Student.NAME,StudentCourse.COURSE_ID 
FROM Student
FULL JOIN StudentCourse 
ON StudentCourse.ROLL_NO = Student.ROLL_NO;
```
**MYSQL does not natively support the `FULL OUTER JOIN`. Instead, you can achieve the same result using a combination of `LEFT JOIN, RIGHT JOIN, UNION` OPERATORS. 

```sql
SELECT * FROM table1
LEFT JOIN table2 ON table1.column_name = table2.column_name
UNION
SELECT * FROM table1
RIGHT JOIN table2 ON table1.column_name = table2.column_name;
```

# <center>5. SELF JOIN</center>
`SELF JOIN` is a regular join, but the table is joined with itself.

**SYNTAX**
```sql
SELECT column_name(s)
FROM table1 T1, table1 T2
WHERE condition;
-- T1 and T2 are different table aliases for the same table.

SELECT alias1.column, alias2.column
FROM table_name alias1
JOIN table_name alias2 ON alias1.related_column = alias2.related_column;
```
**Always use `aliases` in self joins to avoid ambiguity and improve readability. For large datasets ensure proper indexing on the join columns for better performance.**

**Common use cases:**
1. Finding hierarchical relationships (e.g., employees and managers).
2. Comparing rows in the same table (e.g., products with the same price).
3. Detecting duplicates.

```sql
-- Self join to get employees with their managers
SELECT e.Name AS Employee, m.Name AS Manager
FROM Employees e
JOIN Employees m ON e.ManagerID = m.EmployeeID;

SELECT A.CustomerName AS CustomerName1, B.CustomerName AS CustomerName2, A.City
FROM Customers A, Customers B
WHERE A.CustomerID <> B.CustomerID
AND A.City = B.City
ORDER BY A.City;
```

# <center>SET OPERATORS</center>
* Combine the results of multiple queries into a single result set.

![s1](../Resources/assets/joins/s1.png)

## SET RULES
1. `set` operators can be used in any clause.
2. `ORDER BY` is allowed only once - at the end of the query.
3. Each query must have the same number of columns.
4. Column data types must be compatible across queries.
5. The result set takes column names from the first query.


![s2](../Resources/assets/joins/s2.png)

1. `UNION`: Returns All `distinct` rows from both tables
2. `UNION ALL`: REturns `ALL` rows, including `duplicates`
3. `EXCEPT(MINUS)`: Return `unique` rows in `1st table` that are not in `2nd Table`
4. `INTERSECT`: Return `Common rows` between two tables.

# <center> SET USE CASE</center>

1. `SET` operators like `UNION` help combine data from multiple tables, making it easier to consolidate information for reporting and analysis. 
![u1](../Resources/assets/joins/u1.png)

2. `SET` operators like `EXCEPT` help detect changes between datasets, making it easier to identify new, updated, or missing records duirng data integration. 
3. `SET` operators like `EXCEPT` help verify data completeness by comparing tables across databases, ensuring no records are missing or mismatched.
