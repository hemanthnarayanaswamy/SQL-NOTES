# <center> SQL Aggregation Analytical Functions </center>

- An aggregate function is a function that performs a calculation on a set of values, and returns a single value.

- Aggregate functions are often used with the `GROUP BY` clause of the `SELECT` statement. The `GROUP BY` clause splits the result-set into groups of values and the **aggreagate function** can be used to return a single value for each group. 

1. `MIN()` - returns the smallest value of a column
2. `MAX()` - returns the largest value of a column
3. `COUNT()` - returns the number of rows in a set
4. `SUM()` - returns the sum of a numerical column
5. `AVG()` - returns the average value of a numerical column

![a1](../Resources/assets/sql_functions/aggregate/a1.png)

**Aggregate functions ignore null values (except for `COUNT(*)`).**

```sql
-- Here we use the MIN() function and the GROUP BY clause, to return the smallest price for each category in the Products table:
SELECT MIN(Price) AS HighestPrice, CategoryID
FROM Products
GROUP BY CategoryID;

-- Here we use the MAX() function and the GROUP BY clause, to return the highest price for each category in the Products table:
SELECT MAX(Price) AS HighestPrice, CategoryID
FROM Products
GROUP BY CategoryID;
```

- `COUNT(*)` - Counts the total number of rows in a table (including NULL values).
- `COUNT(columnname)` - Counts all non-null values in the column.
- `COUNT(DISTINCT columnname)` - Counts only the unique, non-null values in the column.

```sql
-- Return the number of records for EACH category in the "Products" table:
SELECT COUNT(*) AS [Number of records], CategoryID
FROM Products
GROUP BY CategoryID;
```

- `SUM(columnname)` - Function ignores all `NULL` values in the column.

```sql
--  Return the Quantity for EACH OrderID in the "OrderDetails" table:
SELECT OrderID, SUM(Quantity) AS [Total Quantity]
FROM OrderDetails
GROUP BY OrderID;

-- We can also join the "OrderDetails" table with the "Products" table to find the actual price,
SELECT SUM(Price * Quantity)
FROM OrderDetails
LEFT JOIN Products ON OrderDetails.ProductID = Products.ProductID;
```

**The `AVG()` function ignores NULL values in the column.**

```sql
-- we use the AVG() function and the GROUP BY clause, to return the average price for EACH category in the "Products" table:
SELECT AVG(Price) AS AveragePrice, CategoryID
FROM Products
GROUP BY CategoryID;

-- Return all products with a higher price than the average price:
SELECT * FROM Products
WHERE price > (SELECT AVG(price) FROM Products);
```

# <center> SQL WINDOW FUNCTIONS </center>

- The aggregate functions perform calculations across rows and return a single output row. Like an aggregate function, a `window function`. Perform calculations on a specific subset of data, without losing the level of details of rows.
- SQL `window functions` allow performing calculations across a set of rows that are related to the current row, without collapsing the result into a single value. They are commonly used for tasks like **aggregates, rankings and running totals.**

## Window Function Syntax

![a7](../Resources/assets/sql_functions/aggregate/a7.png)

```sql
SELECT column_name1, 
       window_function(column_name2) 
       OVER ([PARTITION BY column_name3] [ORDER BY column_name4]) AS new_column
FROM table_name;
```
![a5](../Resources/assets/sql_functions/aggregate/a5.png)

```sql
AVG(Sales) OVER (PARTITION BY Category ORDER BY OrderDate ROWS UNBOUNDED PRECEDING)
```

1. `WINDOW FUNCTION`: Perform calculations within a window.
2. `FUNCTION/WINDOW EXPRESSION`: Arguments you pass to a function. 
3. `OVER CLAUSE`: Tells SQL that the function used is a window function. 
4. `PARTITION BY`: Divides the result set into partitions (Windows)
5. `ORDER BY`: Sort the data within a window
